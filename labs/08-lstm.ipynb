```python
import pandas as pd

# Load the data
data = pd.read_csv('../output/(37t)สถานีอุตุนิยมวิทยาลำปาง.csv', encoding='utf-8')

# Display the first few rows of the data
data.head()
```

```python
# Handle missing values in SO2 column with linear interpolation
data[' SO2 '] = data[' SO2 '].interpolate(method='linear')

# Display the first few rows after interpolation
data
```

```python
# Drop rows with missing SO2 values
data_cleaned = data.dropna(subset=[' SO2 '])

# Display the first few rows after dropping missing values
data_cleaned.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Extract the SO2 series
so2_series = data_cleaned[' SO2 '].values.reshape(-1, 1)

# Normalize the series
scaler = MinMaxScaler(feature_range=(0, 1))
so2_normalized = scaler.fit_transform(so2_series)

so2_normalized[:5]  # Display the first few normalized values
```

```python
import numpy as np

def create_dataset(dataset, look_back=48, forecast_horizon=48):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - forecast_horizon + 1):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back:i + look_back + forecast_horizon, 0])
    return np.array(dataX), np.array(dataY)

# Splitting data into training and testing sets (80% train, 20% test)
train_size = int(len(so2_normalized) * 0.8)
test_size = len(so2_normalized) - train_size
train, test = so2_normalized[0:train_size,:], so2_normalized[train_size:len(so2_normalized),:]

# Reshape data for LSTM model
look_back = 48
forecast_horizon = 24*30
trainX, trainY = create_dataset(train, look_back, forecast_horizon)
testX, testY = create_dataset(test, look_back, forecast_horizon)

# Reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))

trainX.shape, trainY.shape, testX.shape, testY.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout


gpus = tf.config.list_physical_devices('GPU')
if gpus:
  # Restrict TensorFlow to only use the first GPU
  try:
    tf.config.set_visible_devices(gpus[0], 'GPU')
    logical_gpus = tf.config.list_logical_devices('GPU')
    print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPU")
  except RuntimeError as e:
    # Visible devices must be set before GPUs have been initialized
    print(e)

# Define the LSTM model
model = Sequential()
model.add(LSTM(100, input_shape=(look_back, 1), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dense(forecast_horizon))
# model.add(Dropout(0.2))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
history = model.fit(trainX, trainY, epochs=10, batch_size=2048, validation_data=(testX, testY), verbose=1)
```

```python
import matplotlib.pyplot as plt
# Predict the next 2 days of SO2 values for the test data
predicted_values = model.predict(testX)

# Inverse transform the normalized predictions to get the actual SO2 values
predicted_SO2 = scaler.inverse_transform(predicted_values)

# Inverse transform the testY values to get the actual SO2 values for comparison
actual_SO2 = scaler.inverse_transform(testY)

# Plot the results
plt.figure(figsize=(20, 8))
plt.plot(actual_SO2[:, 0], label="Actual SO2 Values", color="blue", linestyle='-')
plt.plot(predicted_SO2[:, 0], label="Predicted SO2 Values", color="red", alpha=0.7, linestyle='-')
plt.title("SO2 Prediction for the Next 2 Days")
plt.xlabel("Time")
plt.ylabel("SO2 Value")
plt.legend()
plt.savefig('predict.png')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Make predictions
predicted_values = model.predict(testX)

# Inverse transform the predictions and actuals
predicted_SO2 = scaler.inverse_transform(predicted_values)
actual_SO2 = scaler.inverse_transform(testY)

# Compute metrics
mae = mean_absolute_error(actual_SO2, predicted_SO2)
mse = mean_squared_error(actual_SO2, predicted_SO2)
rmse = np.sqrt(mse)

mae, rmse
```
