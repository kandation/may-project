```python
import os
import pandas as pd

# Directory list
list_of_dirs = [
    'BP', 'DEV', 'NO2_2009-2019', 'PM10_2009-2019', 'RAIN', 'RH',
    'SO2_2009', 'SO2_2010-2020', 'SW', 'TEMP', 'TSP_2009-2019', 'WD', 'WS'
]

# Prepare a dictionary for merging directories with similar prefixes
dir_map = {
    'BP': ['BP'],
    'DEV': ['DEV'],
    'NO2': ['NO2_2009-2019'],
    'PM10': ['PM10_2009-2019'],
    'RAIN': ['RAIN'],
    'RH': ['RH'],
    'SO2': ['SO2_2009', 'SO2_2010-2020'],
    'SW': ['SW'],
    'TEMP': ['TEMP'],
    'TSP': ['TSP_2009-2019'],
    'WD': ['WD'],
    'WS': ['WS']
}

master_data = []

# Iterate over the directory map
for prefix, dirs in dir_map.items():
    for dirx in dirs:
        directory = os.path.join('outputs', dirx)
        files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.csv')]

        for file_path in files:
            df = pd.read_csv(file_path)
            # Add prefix to column names except for 'datetime', 'date', and 'hour'
            df.columns = [col if col in ['datetime', 'date', 'hour'] else prefix + "_" + col for col in df.columns]
            master_data.append(df)

# Concatenate all dataframes
master_df = pd.concat(master_data, axis=0, ignore_index=True)

# Save the master dataframe to a new CSV file
master_df
```

```python
import os
import pandas as pd

# Placeholder for the list of dataframes to be merged
dfs = []

for dirx in list_of_dirs:
    # Extract the base pollutant name (prefix before underscore)
    prefix = dirx.split('_')[0]

    directory = os.path.join('outputs', dirx)
    files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.csv')]

    for file_path in files:
        df = pd.read_csv(file_path)

        # Rename the 'Value' column to the prefix and drop the 'ValueLn' column
        df[prefix] = df['Value']
        df = df.drop(columns=['Value', 'ValueLn'])

        dfs.append(df)

# Merge all dataframes on 'Year' and 'Datetime'
final_df = dfs[0]
for df in dfs[1:]:
    final_df = pd.merge(final_df, df, on=['Year', 'Datetime'], how='outer')

# Save the merged dataframe
final_df
```

```python
import os
import pandas as pd

# Dictionary to hold dataframes
dfs = {}

for dirx in list_of_dirs:
    # Extracting the column name from directory name
    if "_" in dirx:
        col_name = dirx.split("_")[0]
    else:
        col_name = dirx

    # list csv files in the directory
    directory = os.path.join('outputs', dirx)
    files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.csv')]

    # Read and merge files from the same directory
    dir_dfs = [pd.read_csv(file) for file in files]
    dir_df = pd.concat(dir_dfs, ignore_index=True)

    # If the column name already exists, we sum the data, else we just assign
    if col_name in dfs:
        dfs[col_name] = dfs[col_name].add(dir_df, fill_value=0)
    else:
        dfs[col_name] = dir_df

# Joining on 'Year,Datetime'
final_df = dfs.pop('Year,Datetime')  # Start with 'Year,Datetime' dataframe
for col_name, df in dfs.items():
    final_df = final_df.merge(df[['Year', 'Datetime', col_name]], on=['Year', 'Datetime'], how='outer')

# Save the merged dataframe
final_df
```
