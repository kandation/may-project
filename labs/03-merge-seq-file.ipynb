# Make Column Data in each dir

```python
from pathlib import Path
from pprint import pprint


list_of_dirs = path_of_out_value.glob('*/')

dir_renames = {'BP': 'BP',
 'DEV': 'DEV',
 'NO2_2009-2019': 'NO2',
 'PM10_2009-2019': 'PM10',
 'RAIN': 'RAIN',
 'RH': 'RH',
 'SO2_2009': 'SO2_2009',
 'SO2_2010-2020':'SO2' ,
 'SW':'SW',
 'TEMP':'TEMP',
 'TSP_2009-2019':'TSP',
 'WD':'WD',
 'WS':'WS'}

dir_names = set()

# Dictionary to store the result
file_locations = {}
file_loc = {}
for dirx in list_of_dirs:
    # Ensure it's a directory
    if dirx.is_dir():
        # Iterate over CSV files in the directory
        csv_files = dirx.glob('*.csv')

        for file in csv_files:
            # Get file name
            file_name = file.name
            # Prepare the dictionary entry
            file_info = {'dir': dir_renames[dirx.name], 'path': str(file)}

            # Add the file info to the dictionary
            if file_name not in file_locations:
                file_locations[file_name] = []
                file_loc[file_name] = set()
            file_locations[file_name].append(file_info)
            file_loc[file_name].add(dir_renames[dirx.name])


# Print the result
pprint(file_loc)



```

# Visualize available data


```python
import pandas as pd

data = file_locations
all_dirs = {d['dir'] for dirs in data.values() for d in dirs}


rows = []
priority_cols = ['SO2']

for file, dirs in data.items():
    # Create a row with all False values
    row = {dir: False for dir in all_dirs}
    row['CSV File Name'] = file

    # Set True for directories present in this file
    for d in dirs:
        row[d['dir']] = True

    rows.append(row)


df = pd.DataFrame(rows)

df = df[['CSV File Name'] + [col for col in df.columns if col != 'CSV File Name']]

df.replace({True: '✓', False: ''}, inplace=True)

other_cols = sorted([col for col in df.columns if col not in priority_cols and col != 'CSV File Name'])

df = df[['CSV File Name'] + priority_cols + other_cols]
# Add a new column for the sort key (number of check marks in non-priority columns)
df['sort_key'] = df[other_cols].apply(lambda x: x.str.count('✓').sum(), axis=1)

# Sort the DataFrame first by priority columns (descending), then by sort key (descending)
df.sort_values(by=priority_cols + ['sort_key'], ascending=[False, False], inplace=True)

# Drop the sort key column as it's no longer needed
df.drop(columns=['sort_key'], inplace=True)


df
```

# Merge all data each file into columns

```python
# Merge the files
import pandas as pd



for file_name, infoes in file_locations.items():
    merged_data = pd.DataFrame()
    for info in infoes:
        print(info)
        topic = info.get('dir','unk')
        dirs = info.get('path')

        # Read the CSV file
        df = pd.read_csv(dirs)

        df_value = df[['Datetime', 'Value', 'ValueLn']]
        df_value.rename(columns={'Value': f'Value_{topic}', 'ValueLn': f'ValueLn_{topic}'}, inplace=True)

        if merged_data.empty:
                merged_data = df_value
        else:
            merged_data = pd.merge(merged_data, df_value, on='Datetime', how='outer')
    
    print(merged_data.head())
    out_csv = path_of_output / 'merged' / f'merged_{file_name}.csv'
    out_csv.parent.mkdir(parents=True, exist_ok=True)
    merged_data.to_csv(out_csv, index=False)
```
