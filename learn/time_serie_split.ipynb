```python
import numpy as np
from matplotlib import pyplot
from sklearn.model_selection import TimeSeriesSplit

# Sample data
X = np.array(range(100))
y = np.array(range(0, 1000, 10))

# TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

# Plot configuration
pyplot.figure(figsize=(15, 8))

# Plot each split
for index, (train_index, test_index) in enumerate(tscv.split(X), start=1):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Subplot for each split
    pyplot.subplot(5, 1, index)
    pyplot.plot(X_train, y_train, label='Training')
    pyplot.plot(X_test, y_test, label='Test', linestyle='--')
    pyplot.legend()

pyplot.tight_layout()
pyplot.show()
```
